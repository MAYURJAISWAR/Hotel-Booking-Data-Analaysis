# Data Analysis and Visualizations

```python
cancelled_percent=hotel_booking_dataset_copy['is_canceled'].value_counts(normalize=True)
```

```python
cancelled_percent
```

```python
plt.figure()
plt.title('Reservation status count')
plt.bar(['Not cancelled','Cancelled'],hotel_booking_dataset_copy['is_canceled'].value_counts(),edgecolor='k')
plt.show()
```

```python
plt.figure()
ax1=sns.countplot(x='hotel',hue='is_canceled',data=hotel_booking_dataset_copy,edgecolor='k')
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels')
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
```

```python
resort_hotel=hotel_booking_dataset_copy[hotel_booking_dataset_copy['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=hotel_booking_dataset_copy[hotel_booking_dataset_copy['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate in City and Resort Hotel')
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend()
plt.show()
```

```python
hotel_booking_dataset_copy['month']=hotel_booking_dataset_copy['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=hotel_booking_dataset_copy,edgecolor='k')
#legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month')
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot('month','adr',data=hotel_booking_dataset_copy[hotel_booking_dataset_copy['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data=hotel_booking_dataset_copy[hotel_booking_dataset_copy['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with highest reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
hotel_booking_dataset_copy['market_segment'].value_counts()
```

```python
hotel_booking_dataset_copy['market_segment'].value_counts(normalize=True)
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = hotel_booking_dataset_copy[hotel_booking_dataset_copy['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```
